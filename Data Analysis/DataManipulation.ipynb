```python
# Filling the nan values with the mean of the column 
df['Satisfaction_Rating_filling_NA'] = df['Satisfaction_Rating'].fillna(df['Satisfaction_Rating'].mean())
```

```python
df
```

```python
# Renaming column
# df.dtypes
df = df.rename(columns={'Customer_ID':'ID'})
df.head()
```

```python
# Changing data type 
df['Age'] = df['Age'].astype(str)
df.head()
```

```python
# Applying a function 
df['New'] = df['Age'].apply(lambda x:x**2)
df.head()
```

```python
# Aggregation 
grouped_mean = df.groupby('Brand_Purchased')['Satisfaction_Rating'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Brand_Purchased','City'])['Satisfaction_Rating'].sum()
print(grouped_sum)
```

```python
# Aggregate multiple functions 
grouped_agg = df.groupby('City')['Satisfaction_Rating'].agg(['mean', 'sum', 'count'])
print(grouped_agg)
```

```python
# Merging and joining data frames 
df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'], 'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merge data frame on the key columns 
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
